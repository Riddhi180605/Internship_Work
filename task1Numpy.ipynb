**4. Flatten vs Ravel Logic**

Create a 3×3 matrix.

Use ravel() to flatten and modify the first element.

Then check whether original array changes.

Explain why.

```python
arr = np.arange(1,10).reshape(3,3)
arr1 = arr.ravel()
arr1[0] = 100
print("Ravel : \n " ,arr)
arr1 = arr.flatten()
arr1[1] = 100
print("Flatten : \n " ,arr)

# ravel() returns a view of the original array, so modifying it affects the original array, whereas flatten() returns a copy.
```

**10. Multi-step Transformation**

Create array from 1–16.

Steps:

Reshape to 4×4

Multiply each column by [1,2,3,4]

Flatten result

Find standard deviation

```python
arr = np.arange(1,17).reshape(4,4)
print("Array is : \n",arr)
print("\n")
arr1 = arr*np.arange(1,5)
print("Array after multiplication : \n",arr1)
print("\n")
arr2 = arr1.flatten()
print("Array after flattening : \n",arr2)
print("\n")
print("Standard Deviation : ",np.std(arr2))
```

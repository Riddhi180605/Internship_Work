```python
import numpy as np
```

```python
x = np.random.randint(1,20,(3,3,3))
x
```

```python
print(np.min(x))
print(np.min(x,axis = 1))
print(np.min(x,axis = 0))
```

```python
print(np.max(x))
print(np.max(x,axis = 1))
print(np.max(x,axis = 0))
```

```python
print(np.mean(x))
print(np.mean(x,axis = 1))
print(np.mean(x,axis = 0))
```

```python

ans = np.sum(x)
print(ans)
ans = np.sum(x,axis=0)
print(ans) # col vise
ans = np.sum(x,axis=1)
print(ans) # row vise
```

```python
print(x.shape)
print(x.ndim)
print(x.size)
print(x.dtype)
```

```python
x = np.zeros((5,2,3))
print(x)
```

```python
x = np.ones((2,3,2))
print(x)
```

```python
x = np.full((2,2,3),fill_value= 11)
print(x)
```

```python
x = np.empty(shape = (2,3,2))
print(x)
```

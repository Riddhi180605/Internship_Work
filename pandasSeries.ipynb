```python
import numpy as np
s = pd.Series([10,20,30,0,40,None,50,np.nan,60])
# print(s)

print(s.isna())
print(s.isna().sum())
print(s.notnull())
print(s.notnull().sum())
```

```python
import pandas as pd

s = pd.Series([10,20,30,40])
# print(s)

#apply
s = s.apply(lambda x: x*2)
print(s)





```

```python
users = pd.Series(["user1","user2","user3","user4"])
# print(users)

users = users.apply(lambda x: x.upper())
print(users)
```

```python
#map

s = pd.Series([10,20,30,40])
# print(s)

s = s.map(lambda x: x*2)
print(s)
```
